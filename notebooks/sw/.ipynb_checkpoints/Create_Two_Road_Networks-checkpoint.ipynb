# Creating two road networks

1. Removing the South Road and its neighboring roads
2. Strengthening the capacity of South Road

```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
data_mount = '/Users/shenhaowang/Dropbox (MIT)/project_media_lab_South_Australia/SA data/'

all_roads = gpd.read_file(data_mount + 'dataSA/roads/Roads_GDA2020.shp')
South_road = gpd.read_file(data_mount + 'dataSA/roads/South_Road_Highway.shp')
```

```python
all_roads_HWY=all_roads.loc[all_roads['class'] == 'HWY', :]
all_roads_rest=all_roads.loc[all_roads['class'] != 'HWY', :]
```

```python
all_roads_HWY_without_south_road = all_roads_HWY.loc[~all_roads_HWY['geometry'].isin(South_road['geometry']), :]
```

```python
# combine
all_roads_rest_without_sr = pd.concat([all_roads_rest, all_roads_HWY_without_south_road], axis=0)
print(all_roads_rest_without_sr.shape)
```

## Save the file

```python
# Save
all_roads_rest_without_sr.to_file(data_mount + 'dataSA/roads/Roads_GDA2020_without_South_Road.shp')
```

```python
# save
all_roads_HWY_without_south_road.to_file(data_mount + 'dataSA/roads/HWY_without_South_Road.shp')
```
